# 무신사 옷이름, 점수, 리뷰 추출

```python

wb=openpyxl.Workbook()
sheet=wb.active
sheet.append(['옷이름','점수','리뷰'])

# selenium에서 사용할 웹 드라이버 절대경로 정보
chromedriver = 'C:\\Users\PC\프로젝트\.conda\chromedriver.exe'
# selenium의 webdriver에 앞서 설치한 chromedriver를 연동한다.
driver=webdriver.Chrome(chromedriver)
driver.maximize_window()
# driver로 특정페이지를 크롤링한다.

driver.get('https://search.musinsa.com/category/001005')
source=driver.page_source
raw=BeautifulSoup(source,'html.parser')
datas=raw.select('#searchList > li')
for data in datas:
    title=data.select_one('div.article_info > p.item_title').text  #브랜드를 옷이름이라고 착각해서 수정해야합니둥..
    url=data.select_one('div.li_inner > div.list_img > a').attrs['href']
    print(title)
    raw=driver.get(url)
    driver.get(url)
    source2=driver.page_source
    raw2=BeautifulSoup(source2,'html.parser')
    
    #리뷰 페이지가 for 문에서 지정한거 보다 적게 있을 경우 에러뜹니다ㅜㅜ
    #try except로 해결해야할듯
    
    for page in range(3,5):
        driver.implicitly_wait(3)
    
        infos=driver.find_elements_by_css_selector('#wrapStyleEstimateList > div > div > div.postRight > div')
        for info in infos:
            score=raw2.select_one('#wrapStyleEstimateList > div > div > div.postRight > div > div.review-size > div > span')
            a=re.findall('\d+',str(score))
            score_=''.join(a)
            review=info.find_element_by_css_selector('div.pContent > div.summary > div > div.pContent_text > span').text
            
            print('점수:',score_,'\n리뷰\n:',review)  
        
        
            sheet.append([title,score_,review])
        #element = WebDriverWait(driver, 30).until(
        #    EC.element_to_be_clickable((By.XPATH, '/*[@id="wrapStyleEstimateList"]/div/div[11]/div[2]/div/a'))
        #)
        #button=driver.find_element_by_xpath('//*[@id="wrapStyleEstimateList"]/div/div[11]/div[2]/div/a[%s]'%page)
        #webdriver.ActionChains(driver).click(button).perform()
        driver.implicitly_wait(3)
        driver.find_element_by_xpath('//*[@id="wrapStyleEstimateList"]/div/div[11]/div[2]/div/a[%s]'%(page+1)).click()
        time.sleep(3)
wb.save('clothes.xlsx')
```

# BeautifulSoup 
- 네이버 약 & 효과 정보 추출

```python

wb=openpyxl.Workbook()
sheet=wb.active
sheet.append(['약이름','효과'])
for page in range(1,10,1):
    # selenium에서 사용할 웹 드라이버 절대경로 정보
    chromedriver = 'C:\\Users\PC\프로젝트\.conda\chromedriver.exe'
    # selenium의 webdriver에 앞서 설치한 chromedriver를 연동한다.
    driver=webdriver.Chrome(chromedriver)
    # driver로 특정페이지를 크롤링한다.
    
    driver.get('https://terms.naver.com/medicineSearch.naver?page='+str(page))
    html=BeautifulSoup(driver.page_source,'html.parser')

    medis=html.select('#content > div.list_wrap > ul > li')
    for medi in medis: 
        medi_name=medi.select_one('div.info_area > div.subject > strong > a')
        url=medi_name.attrs['href']
        print('약 이름:',medi_name.text)
    
        raw=requests.get('https://terms.naver.com'+url)
        each_html=BeautifulSoup(raw.content,'html.parser')
    
    
        datas=each_html.select('#size_ct > p')
        m=datas[3].text
        print('효과\n',m,'\n')
        sheet.append([medi_name.text,m])
wb.save('medicine.xlsx')    
    
    
```
