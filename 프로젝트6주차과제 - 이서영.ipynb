```python
wb=openpyxl.Workbook()
sheet=wb.active
sheet.append(['title','user_id','user_info','score','review'])

# selenium에서 사용할 웹 드라이버 절대경로 정보
chromedriver = 'C:\\Users\PC\프로젝트\.conda\chromedriver.exe'
# selenium의 webdriver에 앞서 설치한 chromedriver를 연동한다.
driver=webdriver.Chrome(chromedriver)
q=1;j=1;k=1
driver.get('https://store.musinsa.com/app/goods/1915051/0#estimateBox')
try:
    a=0 ; b=0; i=3
    while a<2:
        if (i<8 and b<2):
            try:
                
                sor2=driver.page_source
                raw2=BeautifulSoup(sor2,'html.parser')
                title=raw2.select_one('#page_product_detail > div.right_area.page_detail_product > div.right_contents.section_product_summary > span > em').text
                sheet.cell(row=q+1,column=1).value[title]
                q+=1
                driver.implicitly_wait(2)
                buttons=driver.find_element_by_css_selector('#estimate_goods').click() # 일반후기 클릭
                webdriver.ActionChains(driver).click(buttons).perform()
                    
                 
                print('-'*30,i,' page','-'*30)
                
                infos=driver.find_elements_by_css_selector('#wrapEstimateList > div > div > div.postRight > div')
                            
                source=driver.page_source
                raw=BeautifulSoup(source,'html.parser')
                scores=raw.select('#wrapEstimateList > div > div > div.postRight > div')
            except :pass
                            
            try:
                for sc in scores:
                    score_full = sc.select_one('div.review-size > div > span')
                    score=re.findall('\d+',str(score_full))
                    for sc_i in range(0,len(score)):
                        sheet.cell(row=j+1,column=4).value=score[sc_i]
                        j+=1
                                    
                    print(score)
            except :
                Null="Null"
                sheet.cell(row=j+1,column=3).value[Null]
                
                            
                            
                            
            for info in infos:
                               
                    
                                #유저정보와 리뷰 추출
                try:
                    user_id=info.find_element_by_css_selector('div.profile > p > span.gallery_lv > span').text
                except : user_id="Null"
                try:
                    review=info.find_element_by_css_selector('div.pContent > div.summary > div > div.pContent_text > span').text
                except : review="Null"
                            
    
                sheet.cell(row=k+1,column=1).value=user_id
                                
                sheet.cell(row=k+1,column=4).value=review
                                
                print('사용자:',user_id,'\n리뷰\n:',review)
                k+=1
                                
                                
                                
            
                
                
                    
                
                
                            #후기 2페이지 추출
            driver.implicitly_wait(1)
            buttons2=driver.find_elements_by_css_selector('#wrapEstimateList > div > div.nslist_bottom > div.pagination.textRight > div > a')[i]
            webdriver.ActionChains(driver).click(buttons2).perform()
            time.sleep(3)
                    
            i +=1
        elif (i==8 and b<2): 
            print('click right button')
            i=3
            b+=1
        else:
            break
                
        
except:pass
wb.save('a.xlsx')
```
