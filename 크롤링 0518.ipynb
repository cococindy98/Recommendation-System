```python
wb=openpyxl.Workbook()
sheet=wb.active
sheet.append(['옷이름','브랜드','태그','가격','연령','성별','유저아이디','유저정보','평점','리뷰'])

chromedriver =  'C:\\Users\PC\프로젝트\.conda\chromedriver.exe'
# selenium의 webdriver에 앞서 설치한 chromedriver를 연동한다.
driver=webdriver.Chrome(chromedriver)
driver.maximize_window()
# driver로 특정페이지를 크롤링한다.
for page in range(1,4):
    driver.get('https://search.musinsa.com/category/001001'+'?device=&d_cat_cd=001001&brand=&rate=&page_kind=search&list_kind=small&sort=pop&sub_sort=&page='+
                str(page)+'&display_cnt=90&sale_goods=&ex_soldout=&color=&price1=&price2=&exclusive_yn=&size=&tags=&sale_campaign_yn=&timesale_yn=&q=')
        
    source=driver.page_source
    raw=BeautifulSoup(source,'html.parser')
    datas=raw.select('#searchList > li')
        
    for data in datas:
        title=data.select_one('div.article_info > p.list_info > a').attrs['title'] # 옷 이름
        url=data.select_one('div.li_inner > div.list_img > a').attrs['href']      # 상세 옷 주소
        print(title)

        raw=driver.get(url)
        driver.get(url)
        source2=driver.page_source
        raw2=BeautifulSoup(source2,'html.parser')
    
        #상세 페이지에서 개별 옷 브랜드와 품번 수집
        brands=raw2.select('#product_order_info > div.explan_product.product_info_section > ul > li')[0]
        brand_product_num=brands.text[9:]
          
        
        #태그 한글..로 된것만 찾음..문제
        tag_full=raw2.select('#product_order_info > div.explan_product.product_info_section > ul > li.article-tag-list.list > p')
        tag=re.compile('[가-힣]+').findall(str(tag_full))
        
        # 가격
        price_full=raw2.select('#product_order_info > div.explan_product.price_info_section > ul > li:nth-child(2) > p.product_article_contents')
        #price_no_dot=re.sub(r"[^a-zA-Z0-9]","",str(price))
        # 가격 형태 11,900  30,000   정규표현식에서 ',' 있으면 이상하게 추출
        # ',' 제거후에 가격에 해당하는 숫자만 추출
        price=re.findall('\d+',re.sub(r'[^a-zA-Z0-9]','',str(price_full)))[0]
        
               
                
        #연령
        age_full=raw2.select('#graph_summary_area > strong:nth-child(2) > em')
        age=re.findall('\d{2}~\d{2}',str(age_full))
        
        #성별
        gender=raw2.select_one('#graph_summary_area > span.man.graph_sex_text').text
        
        sheet.append([title,brand_product_num,tag,price,age,gender]) 
                
            
    
        
        try:
            a=0 ; b=0; i=3
            while a<2:
                if (i<8 and b<2):
                
                    driver.implicitly_wait(3)
                    buttons=driver.find_element_by_css_selector('#estimate_goods').click() # 일반후기 클릭
                    webdriver.ActionChains(driver).click(buttons).perform()
                    
                    
                    print('-'*30,i,' page','-'*30)
                    
                    infos=driver.find_elements_by_css_selector('#wrapEstimateList > div > div > div.postRight > div')
                    source=driver.page_source
                    raw=BeautifulSoup(source,'html.parser')
                    scores=raw.select('#wrapEstimateList > div > div > div.postRight > div')
                    for sc in scores:
                        score_full = sc.select('#wrapEstimateList > div > div > div.postRight > div > div.review-size > div > span')
                        score=re.findall('\d+',str(score_full))
                        print(score)
                    for info in infos:
                        print(1)
                    
                        #유저정보와 리뷰 추출
                        user=info.find_element_by_css_selector('div.profile > p > span.gallery_lv > span').text
                        user_info=info.find_element_by_css_selector('div.connect_product.estimate-item > div.connect_review_info > p').text
                        review=info.find_element_by_css_selector('div.pContent > div.summary > div > div.pContent_text > span').text
                        
                        
    
                        #score=re.findall('\d{1}.\d{1}',str(score_full))
                        
                        print('사용자:',user,'\n사용자정보\n:',user_info,'\n리뷰\n:',review)
                        sheet.append([user,user_info,score,review])
                
                    
                
                
                    #후기 2페이지 추출
                    driver.implicitly_wait(3)
                    buttons2=driver.find_elements_by_css_selector('#wrapEstimateList > div > div.nslist_bottom > div.pagination.textRight > div > a')[i]
                    webdriver.ActionChains(driver).click(buttons2).perform()
                    time.sleep(3)
                    
                    i +=1
                elif (i==8 and b<2): 
                    print('click right button')
                    i=3
                    b+=1
                else:
                    break
                
        
        except:pass
wb.save('clothes.xlsx')
```
